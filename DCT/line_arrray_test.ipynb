```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io

def build_line_array(theta, W_D, N):
    n_theta = np.radians(theta)

    # Create a 100x100 array of zeros
    array = np.zeros((N, N))
    
    # Calculate the slope of the line
    slope = np.tan(n_theta)
    
    # Iterate over the rows of the array
    for i in range(N):
        # Calculate the x-coordinate of the line at row i
        x = int(i * slope)
        
        # Add ones to the cells around the calculated x-coordinate
        # Limit x to be within the array bounds
        if 0 <= x < N:
            for w in range(-W_D, W_D + 1):
                if 0 <= x + w < N:
                    array[i, x + w] = 1
    
    return array

OMG = build_line_array(45, 100, 5000)
```

```python
plt.imshow(OMG)
plt.show()
```
